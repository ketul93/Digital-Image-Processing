```python
def my1dfft(signal):
    M = signal.shape[0]
    F = np.zeros(M,dtype=np.complex)
    for i in range(1,(M+1)):
        for k in range(1,M+1):
            F[i-1] += signal[k-1] * (np.exp(-2j*np.pi*(i-1)*(k-1)/M))
    return F
x = np.random.random(1024)
np.allclose(my1dfft(x), np.fft.fft(x))
```
